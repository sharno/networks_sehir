```python
import networkx as nx
import network2
import matplotlib.pyplot as plt

%load_ext autoreload
%autoreload 2
```

```python
G = network2.SocialGraph()
tmp = G.load_data("likes.json", num_posts=None)
```

```python
posts_graph = G.create_posts_graph()
```

```python
nx.info(G.posts_graph)
```

```python
import json
with open("data.json", "r") as f:
    posts_data = json.load(f)

G.posts = {}
for post in posts_data:
    G.posts[post['id']] = post
    del G.posts[post['id']]['id']
```

```python
G.calculate_posts_centralities()
```

# Clustering coefficients

```python
nodes_clustering = nx.clustering(G.posts_graph)
graph_average_clustering = nx.average_clustering(G.posts_graph)

print "Average clustering of the whole graph = ", graph_average_clustering
print "top 10 nodes considering clustering coefficient: "
top_clustering = sorted(nodes_clustering.items(), key=lambda x: x[1], reverse=True)[:10]
```

# Centralities

```python
print "Top 10 highest degree centralities:"
top_degree = sorted(G.posts_graph_data['degree_centrality'].items(), key=lambda x: x[1], reverse=True)[:10]
top_degree = [(x[0], G.posts[x[0]], x[1]) for x in top_degree]
for node in top_degree:
    print "    -", node[1]['message'], " (", node[0], "): ", node[2]
    print
```

```python
print "Top 10 highest closeness centralities:"
top_closeness = sorted(G.posts_graph_data['closeness_centrality'].items(), key=lambda x: x[1], reverse=True)[:10]
top_closeness = [(x[0], G.posts[x[0]], x[1]) for x in top_closeness]
for node in top_closeness:
    print "    -", node[1]['message'], " (", node[0], "): ", node[2]
    print
```

```python
print "Top 10 highest betweenness centralities:"
top_betweenness = sorted(G.posts_graph_data['betweenness_centrality'].items(), key=lambda x: x[1], reverse=True)[:10]
top_betweenness = [(x[0], G.posts[x[0]], x[1]) for x in top_betweenness]
for node in top_betweenness:
    print "    -", node[1], " (", node[0], "): ", node[2]
    print
```

```python
print "Top 10 highest eigenvector centralities:"
top_eigenvector = sorted(G.posts_graph_data['eigenvector_centrality'].items(), key=lambda x: x[1], reverse=True)[:10]
top_eigenvector = [(x[0], G.posts[x[0]], x[1]) for x in top_eigenvector]
for node in top_eigenvector:
    print "    -", node[1], " (", node[0], "): ", node[2]
    print
```

```python
print "Top 10 highest pagerank:"
top_pagerank = sorted(G.posts_graph_data['pagerank'].items(), key=lambda x: x[1], reverse=True)[:10]
top_pagerank = [(x[0], G.posts[x[0]], x[1]) for x in top_eigenvector]
for node in top_pagerank:
    print "    -", node[1], " (", node[0], "): ", node[2]
    print
```

```python

import powerlaw
import numpy
plt.figure(figsize=(10,10))
degree_frequencies = nx.degree_histogram(G.posts_graph)
fit = powerlaw.Fit(numpy.array(degree_frequencies)+1,xmin=1,discrete=True)
fit.power_law.plot_pdf( color= 'b',linestyle='--',label='fit ccdf')
fit.plot_pdf( color= 'b')
plt.show()
print('alpha= ',fit.power_law.alpha,'  sigma= ',fit.power_law.sigma)
```
